**Importing Dependencies**

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_India'

page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
```

```python
table = soup.find_all('table')[0]
```

```python
titles = table.find_all('th')
```

```python
india_titles = [title.text.rstrip() for title in titles]

india_titles
```

**Building the DataFrame**

```python
df = pd.DataFrame(columns = india_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
column_names = ['Column1', 'Column2', 'Column4']
for row in column_data:
    row_data = row.find_all('td')
    individual_row_data = [data.text.rstrip() for data in row_data]

    # Removing columns at index 1 and 3
    individual_row_data = individual_row_data[:1] + individual_row_data[2:3]+ individual_row_data[4:]

   # Check if the modified row has the same number of elements as the columns
    if len(individual_row_data) == len(column_names):
        # Append the modified row to the DataFrame
        df.loc[len(df)] = individual_row_data
```

```python
df
```

**Saving the file**

```python
df.to_csv(r'Top 50 Companies.csv', index = False)
```
